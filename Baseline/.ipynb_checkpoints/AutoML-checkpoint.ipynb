## 분류
* https://pycaret.org/classification1/
* https://pycaret.readthedocs.io/en/latest/api/classification.html

```python
from pycaret.classification import *
```

```python
from pycaret.datasets import get_data
train = get_data('juice')
```

```python
train.Purchase = pd.factorize(train.Purchase)[0]
train.Store7 = pd.factorize(train.Store7)[0]
```

```python
clf = setup(session_id=42, data=train, target='Purchase')
```

```python
best = compare_models(fold = 5, round = 3, sort = 'AUC', n_select = 5)
```

```python
best
```

```python
# https://dacon.io/codeshare/2428

blended = blend_models(estimator_list = best, fold = 5, method = 'soft')
```

```python
pred_holdout = predict_model(blended)
```
