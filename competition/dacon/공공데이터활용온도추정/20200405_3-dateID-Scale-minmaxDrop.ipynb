```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
submission = pd.read_csv('data/sample_submission.csv')

train.head(3)
train.tail(3)
```

```python
train['date'] = pd.PeriodIndex(start='2020-01-01', end=None, periods=len(train), freq='10Min')
test['date'] = pd.PeriodIndex(start='2020-02-03', end=None, periods=len(test), freq='10Min')
train.tail()
```

```python
train.set_index('date', drop=False, inplace=True)
test.set_index('date', drop=False, inplace=True)
test.tail()
```

```python
#train.drop(['id'], axis=1, inplace=True)
#test.drop(['id'], axis=1, inplace=True)
```

```python
train.head()
```

```python
# 평균기온
mean_temp = train.loc[:,"Y00":"Y18"].mean(axis=1)
mean_temp
```

```python
t = train["Y18"].isna() 
null_index = t[t==True].index
# 평균기온으로 Y18 매움
train.loc[null_index, "Y18"] = train.loc[null_index, "Y00":"Y18"].mean(axis=1)
```

```python
train.head()
```

```python
def standardization(df):
    mean = np.mean(df)
    std = np.std(df)
    norm = (df - mean) / (std - 1e-07)
    return norm, mean, std
```

```python
X_name = train.loc[:,"X00":"X39"].columns
X_name
```

```python
train_X_norm, mean, std = standardization(train[X_name]) 

#테스트용 데이터 표준화
test_X_norm = (test[X_name] - mean) / (std - 1e-07)
```

```python
train_X_norm.shape, train.shape
```

```python
train['date']
```

```python
train2 = pd.concat([train['date'], train_X_norm], axis=1) 
test2  = pd.concat([test['date'] , test_X_norm] , axis=1)

print(train2.shape, test2.shape)
```

```python
train2.head()
```

```python
test2.head()
```

```python
def same_min_max(df):
    return df.drop(df.columns[df.max() == df.min()], axis=1)

train_new = same_min_max(train)
test_new  = same_min_max(test)
train_new = same_min_max(train2)
test_new  = same_min_max(test2)
```

```python
import lightgbm as lgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
train2.drop(['date'], axis=1, inplace=True)
```

```python
train2.head()
```

```python
train['Y18']
```

```python
lgb_train = lgb.Dataset(train2, label=train['Y18'])

lgb_param = {
    "objective":"regression",
    "metrics":"mae",
    "learning_rate":0.01
}

print("cv start")
cv_result = lgb.cv(
    lgb_param,
    lgb_train,
    num_boost_round=99999,
    nfold=5,
    early_stopping_rounds=10,
    stratified=False,
    verbose_eval=10 
)

print("train start")
lgb_model = lgb.train(
    lgb_param,
    lgb_train,
    num_boost_round=len(cv_result["l1-mean"])
)
```

```python
test2.drop(['date'], axis=1, inplace=True)
test2.head()
```

```python
#submission["Y18"] = lgb_model.predict(test.loc[:, "X00":"X39"])
submission["Y18"] = lgb_model.predict(test2)
submission.head()
```
