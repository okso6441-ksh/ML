```python
def value_counts(df, cols):
    for col in cols:
        print(col)
        print(df[col].value_counts(),'\n')
```

```python
def importance_plotting(data, xlabel, ylabel, n=20):
    sns.set(style="whitegrid")
    ax = data.tail(n).plot(kind='barh')
    
    ax.set(xlabel=xlabel, ylabel=ylabel)
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)
    plt.show()
```
